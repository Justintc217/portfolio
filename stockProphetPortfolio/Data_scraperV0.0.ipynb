```python
def gather_minutes(tics, start, finish, stopping = True):
#     global dfs
#     global save_dfs
#     save_dfs.append(list(dfs))
    dfs = []
    for tic in tics[start:finish]:
        t = tics.index(tic)
        print(t , tic)
        df = alphav_stockdata(tic, sound = True)
        if type(df) != str:
            dfs.append(df)
        else:
            if stopping is True:
                print("WARNING THE PROGRAM HAS STOPPED error with: {}{}{:5}".format("--------/n"*2, tics.index(df), df))
                return dfs
            else:
                print("WARNING, MOVING TO NEXT TICKER, error with: {}{}{:5}".format("--------/n"*2, tics.index(df), df))
    return dfs
```
